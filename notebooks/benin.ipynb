OUTLIER DETECTION

```python
# identify numeric columns
numeric_cols = df.select_dtypes(include="number").columns
print(f"Numeric columns:{numeric_cols}")
```

```python
# compute Z-scores and flag outliers(|Z| > 3)
z_scores=np.abs(stats.zscore(df[numeric_cols]))
outlier_flags = (z_scores > 3).any(axis=1)
print(f"Number of outlier samples flagged: {outlier_flags.sum()}")
```

```python
outlier_flags = pd.DataFrame(index=df.index)

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_flags[col] = ~df[col].between(lower_bound, upper_bound)

outlier_summary = outlier_flags.sum().sort_values(ascending=False)
print("Outliers per column:\n", outlier_summary)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.plot(df.index, df['WSstdev'], label='WSstdev')
plt.title('Time Series of WSstdev')
plt.xlabel('Time')
plt.ylabel('WSstdev')
plt.legend()
plt.show()
```

```python
outliers_ws = outlier_flags['WSstdev']

plt.figure(figsize=(15, 5))
plt.plot(df.index, df['WSstdev'], label='WSstdev', alpha=0.7)
plt.scatter(df.index[outliers_ws], df['WSstdev'][outliers_ws], color='red', label='Outliers', s=10)
plt.title('WSstdev with Outliers Highlighted')
plt.xlabel('Time')
plt.ylabel('WSstdev')
plt.legend()
plt.show()
```

```python
# Extract rows where WSstdev is flagged as outlier
ws_outlier_indices = df.index[outliers_ws]

# Look at WS values during those intervals
ws_outliers = df.loc[ws_outlier_indices, 'WS']

print(ws_outliers.describe())  # Summary stats of wind speed during WSstdev outliers

# Optionally, plot raw wind speed at those times
plt.figure(figsize=(15, 5))
plt.plot(df.index, df['WS'], label='Wind Speed')
plt.scatter(ws_outlier_indices, ws_outliers, color='orange', label='WSstdev Outlier Times', s=10)
plt.title('Wind Speed with WSstdev Outlier Times Highlighted')
plt.xlabel('Time')
plt.ylabel('Wind Speed')
plt.legend()
plt.show()
```

```python
# Define threshold for "steady" or "low" WS — adjust based on your data
steady_ws_threshold = 2.0  # Example: 2.0 m/s or less considered steady

# Flag WSstdev outliers where WS is steady (possible false outliers)
false_outliers = (outlier_flags['WSstdev']) & (df['WS'] <= steady_ws_threshold)

# Flag WSstdev outliers where WS is variable/high (likely true outliers)
true_outliers = (outlier_flags['WSstdev']) & (df['WS'] > steady_ws_threshold)

print(f"Number of likely false outliers: {false_outliers.sum()}")
print(f"Number of likely true outliers: {true_outliers.sum()}")
```

```python
from scipy.stats import zscore

# Create a DataFrame to store Z-scores
z_scores = df[numeric_cols].apply(lambda x: zscore(x.fillna(x.mean())))

# Flag outliers where absolute Z-score > 3
outlier_flags_z = (z_scores.abs() > 3)

# Count outliers per column
outlier_counts_z = outlier_flags_z.sum()

print("Number of Z-score outliers per column:")
print(outlier_counts_z)
```

```python
# 1. Calculate Z-score outliers
z_scores = df[numeric_cols].apply(lambda x: zscore(x.fillna(x.mean())))
outlier_flags_z = (z_scores.abs() > 3)

# 2. Calculate IQR outliers
outlier_flags_iqr = pd.DataFrame(index=df.index)
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_flags_iqr[col] = ~df[col].between(lower_bound, upper_bound)

# 3. Combine the flags: flag as outlier if flagged by either method
combined_outliers = outlier_flags_z | outlier_flags_iqr

# 4. Count combined outliers per column
combined_outlier_counts = combined_outliers.sum()

print("Number of combined outliers per column:")
print(combined_outlier_counts)
```
