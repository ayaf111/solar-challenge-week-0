## âœ… Winsorization (Outlier Handling)

**Purpose:**  
Reduce the influence of extreme outlier values by capping them at the 1st and 99th percentiles.

**Steps:**
1. Select all numeric columns in the dataset.
2. For each column:
   - Calculate the 1st and 99th percentiles.
   - Replace values below the 1st percentile with the 1st percentile.
   - Replace values above the 99th percentile with the 99th percentile.

**Advantages:**
- Keeps the dataset size the same.
- Reduces the influence of extreme values.
- More robust than deletion and less distortion than mean imputation.

IMPUTATION


```python
# Step 1: Select numeric columns only (excluding Timestamp if present)
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Step 2: Apply Winsorization to each numeric column
for col in numeric_cols:
    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)
    df[col] = np.clip(df[col], lower, upper)
```

```python
sample_df = df.sample(5000, random_state=42)
for col in ['ModA', 'ModB', 'WS']:
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=sample_df[col])
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)
    plt.show()
```

1. WS (Wind Speed)
Median (Q2): The horizontal line inside the box is centered, suggesting a symmetrical distribution of wind speed values.

IQR (Box Height): The box is short, indicating low variability.

Whiskers: Extend close to the box edges, showing most data points lie near the median.

Outliers: Some dots outside the whiskers, representing unusually high or low wind speed values.

Conclusion: Wind speed is fairly stable, with a few anomalies.

2. ModA and ModB (Module Temperature A and B)
Median (Q2): Not centered in the box (closer to the top or bottom), indicating a skewed distribution.

IQR: The box is taller, meaning higher variability in temperature.

Conclusion: ModA and ModB, has wide distribution and skewness, indicating inconsistent thermal behavior across conditions.
